```python
G = net.generator(128, 7, 256)
```

```python
chainer.serializers.load_npz("./result/epoch50.gen", G)
```

```python
z = np.random.random((2, 128)).astype(np.float32)
```

```python
z_new = []
dist = (z[1] - z[0])/16
for i in range(16):
    z_new.append(z[0]+(dist*i))
z_new = np.array(z_new).astype(np.float32)
```

```python
with chainer.using_config("train", False):
    x_fake = G(z)
    x_fake = x_fake.data / 2 + 1
```

```python
width = 2
plt.figure(figsize=(width,width),dpi=200)
for idx, i in enumerate(x_fake):
    plt.subplot(1, width, idx + 1)
    plt.imshow(i[0], cmap="gray")
    plt.xticks(color="None")
    plt.yticks(color="None")
    plt.tick_params(length=0)
```

```python
with chainer.using_config("train", False):
    x_fake = G(z_new)
    x_fake = x_fake.data / 2 + 1
```

```python
width = 16
plt.figure(figsize=(width,width),dpi=200)
for idx, i in enumerate(x_fake):
    plt.subplot(1, width, idx + 1)
    plt.imshow(i[0], cmap="gray")
    plt.xticks(color="None")
    plt.yticks(color="None")
    plt.tick_params(length=0)
```
